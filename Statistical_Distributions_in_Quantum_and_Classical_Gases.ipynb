```python
import matplotlib.pyplot as plt
import numpy as np

e=1.602*10**(-19)     # Electronic charge (C)
kB=1.38*10**(-23)     # Boltzmann constant (J/K)
mu=[0,0,0.56,0,0.35]    # Chemical potential (eV)
T=[300,600,900]

mb_distbn_fn = lambda x,T,mu: 1/(np.exp(((x-mu)*e)/(kB*T)))
be_distbn_fn = lambda x,T,mu: 1/(np.exp(((x-mu)*e)/(kB*T))-1)
fd_distbn_fn = lambda y,T,mu: 1/(np.exp(((y-mu)*e)/(kB*T))+1)



x0=0.01
xn=1
n=100
h=(xn-x0)/n

linestyle=['--','-.',None]
distbn_fn=[mb_distbn_fn, be_distbn_fn, fd_distbn_fn, mb_distbn_fn, fd_distbn_fn]
titles=["Maxwell-Boltzmann of perfect gas, u=0 eV",
        "Bose Einstein for photon gas, u=0 eV",
        "Fermi Dirac for Si, u=0.56 eV",
        "Maxwell-Boltzmann for Hydrogen gas at high temp, u=0 eV",
        "Fermi Dirac for Ge, u=0.35 eV"]

for k in range(5):
    for j in range(3):
        y_coordinate=[]
        x_coordinate=[]

        for i in range(n):
            y_coordinate.append(distbn_fn[k](x0+i*h, T[j], mu[k]))
            x_coordinate.append(x0+i*h)

        plt.plot(x_coordinate, y_coordinate, label=f"T{j}={T[j]}", linestyle = linestyle[j])
    plt.ylabel("f(E) --->")
    plt.xlabel("Energy(eV) --->")
    plt.title(titles[k])
    plt.legend()
    plt.show()
    plt.clf()

labels=["MB","BE","FD"]
for j in range(3):
    y_coordinate=[]
    x_coordinate=[]

    for i in range(n):
        y_coordinate.append(distbn_fn[j](x0+i*h, 900, mu[j]))
        x_coordinate.append(x0+i*h)

    plt.plot(x_coordinate, y_coordinate, label=labels[j], linestyle = linestyle[j])

plt.ylim(0,1.1)
plt.title("Comparison at T=900K")
plt.ylabel("f(E) --->")
plt.xlabel("Energy(eV) --->")
plt.legend()
plt.show()
```
